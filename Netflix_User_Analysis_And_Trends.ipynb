***Import Necessary Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***Load the dataset***

```python
file_path = "/content/drive/MyDrive/netflix_users.csv"
df = pd.read_csv(file_path)
```

*** Converting 'Last_Login' to datetime format***

*   List item
*   List item



```python
df['Last_Login'] = pd.to_datetime(df['Last_Login'])
```

***Aggregate user count by date***

```python
df['Login_YearMonth'] = df['Last_Login'].dt.to_period('M')
user_trend = df.groupby('Login_YearMonth').size().reset_index(name='User_Count')
```

*** Plot Netflix user growth over time***

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=user_trend['Login_YearMonth'].astype(str), y=user_trend['User_Count'], marker="o")
plt.xticks(rotation=45)
plt.xlabel("Year-Month")
plt.ylabel("Number of Active Users")
plt.title("Growth of Netflix Users Over Time")
plt.grid(True)
plt.show()
```

***Analyze demand for different genres***

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df['Favorite_Genre'], order=df['Favorite_Genre'].value_counts().index)
plt.xlabel("Count")
plt.ylabel("Genre")
plt.title("Demand for Different Genres on Netflix")
plt.show()
```

***Watch time distribution per genre***

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Watch_Time_Hours'], y=df['Favorite_Genre'])
plt.xlabel("Watch Time (Hours)")
plt.ylabel("Genre")
plt.title("Watch Time Distribution per Genre")
plt.show()
```
